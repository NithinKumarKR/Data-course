```python
def map_to_month_name(month):
    month_names = ["", "January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
    return month_names[month]
```

```python
from pyspark.sql import functions as F

# 3. Group by "trans_status," "Month," and "Year" and sum "purchase_amount"
trans_type_counts = result_df.groupBy("trans_status", "Month", "Year").agg(
F.regexp_replace(F.format_number(F.sum("purchase_amount"),0).cast("string"), ",", "").alias("distinct_transaction_sum")
)

# Show the final DataFrame with the total row
trans_type_counts.show()
```

```python
trans_type_counts.createOrReplaceTempView("filtered_trans_type_counts")

# Run SQL queries on the registered table
sql_query = """
SELECT
    trans_status,
    Month,Year,distinct_transaction_sum,
    ROUND((distinct_transaction_sum * 100) / (
        SELECT SUM(fc2.distinct_transaction_sum)
        FROM filtered_trans_type_counts fc2
        WHERE fc2.Month = fc.Month
    ), 2)  AS transaction_count
FROM filtered_trans_type_counts fc
"""
# 108698 61727
sql_result = spark.sql(sql_query)

# Show the SQL result
sql_result.show()
```
